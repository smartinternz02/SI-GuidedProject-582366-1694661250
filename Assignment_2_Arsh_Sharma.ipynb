```python
import seaborn as sns 
import matplotlib.pyplot as plt       # ARSH SHARMA 
```

```python
car_crashes = sns.load_dataset("car_crashes")
```

```python
sns.pairplot(car_crashes)
plt.title("Pairplot of Car Crashes Dataset")
plt.show()
```

```python
sns.scatterplot(x="alcohol", y="total", data=car_crashes)
plt.title("Scatterplot of Alcohol vs. Total Crashes")
plt.show()
```

```python
correlation_matrix = car_crashes.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.histplot(car_crashes["speeding"], bins=20, kde=True)
plt.title("Histogram of Speeding")
plt.xlabel("Speeding Incidents")
plt.ylabel("Frequency")
plt.show()
```

```python
selected_states = ['OH', 'WV', 'AL', 'NY', 'NM', 'SC', 'AK', 'AR', 'AZ', 'DE', 'HI', 'ID']
filtered_car_crashes = car_crashes[car_crashes['abbrev'].isin(selected_states)]
sns.boxplot(x="abbrev",y="ins_premium",data=filtered_car_crashes)
plt.xlabel("State Abbreviation")
plt.ylabel("Insurance Premium")
```

```python
selected_states = ['OH', 'WV', 'AL', 'NY', 'NM', 'SC', 'SD', 'TN', 'CA', 'AR', 'AZ', 'CT', 'DC', 'AK']
filtered_car_crashes = car_crashes[car_crashes['abbrev'].isin(selected_states)]
sns.barplot(data=filtered_car_crashes,y="total",x="abbrev",errorbar=None)
plt.xlabel("State Abbreviation")
```

```python
severity_counts = car_crashes['abbrev'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(severity_counts, labels=severity_counts.index, autopct='%0.2f%%', startangle=90)
plt.title('Crash Severity Distribution by State')
plt.show()
```

```python
selected_states = ['OH', 'WV', 'AL', 'NY', 'NM', 'SC', 'SD', 'TN', 'CA', 'AR', 'AZ', 'CT', 'DC', 'AK']
filtered_car_crashes = car_crashes[car_crashes['abbrev'].isin(selected_states)]
sns.lineplot(x="total",y="abbrev",data=filtered_car_crashes,errorbar=None)
plt.xlabel("Total")
plt.ylabel("State Abbreviation")
```

```python
sns.jointplot(x='alcohol', y='total', data=car_crashes, kind='scatter')
plt.show()
```

```python
selected_states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR"]
filtered_car_crashes = car_crashes[car_crashes['abbrev'].isin(selected_states)]
g = sns.relplot(x='alcohol', y='total', data=filtered_car_crashes, height=6, aspect=1.5)
g.set_axis_labels('Alcohol Consumption', 'Total Crashes')
plt.show()
```
